```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# !--------------------- TIME SERIES PLOTTING ---------------------!
def plot_data(data):
    """Plot the data using matplotlib."""
    df = pd.read_csv(data, header=None)
    df.columns = ['timestamp', 'value']
    df['timestamp'] = df['timestamp'].astype(float)
    df['value'] = df['value'].astype(int)
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    
    plt.figure(figsize=(10, 5))
    plt.plot(df['timestamp'], df['value'], marker='o', linestyle='None', markersize=3)
    plt.title('Sensor Data Over Time')
    plt.xlabel('Timestamp (HH:MM:SS)')
    plt.ylabel('Vibration')
    plt.grid(True)
    plt.show()
```

```python
# Coin drop data plots comparing between cloth and table
coin_cloth_data = pd.read_csv('Coin/cloth_30cm/sensor_data_14-56-01_cloth_30cm_3.csv', header=None)
coin_cloth_data.columns = ['timestamp', 'value']
coin_cloth_data['timestamp'] = coin_cloth_data['timestamp'].astype(float)
coin_cloth_data['value'] = coin_cloth_data['value'].astype(int)
coin_cloth_data['timestamp'] = pd.to_datetime(coin_cloth_data['timestamp'], unit='s')

coin_table_data = pd.read_csv('Coin/table_30cm/sensor_data_14-46-16_table_30cm_3.csv', header=None)
coin_table_data.columns = ['timestamp', 'value']
coin_table_data['timestamp'] = coin_table_data['timestamp'].astype(float)
coin_table_data['value'] = coin_table_data['value'].astype(int)
coin_table_data['timestamp'] = pd.to_datetime(coin_table_data['timestamp'], unit='s')

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(coin_cloth_data['timestamp'], coin_cloth_data['value'], marker='o', linestyle='None', markersize=3, label='Cloth')
plt.title('Coin Drop Data Comparison: Cloth vs Table at 30cm Trial 3')
plt.ylabel('Vibration')
plt.subplot(2, 1, 2)
plt.plot(coin_table_data['timestamp'], coin_table_data['value'], marker='o', linestyle='None', markersize=3, label='Table')
plt.xlabel('Timestamp (HH:MM:SS)')
plt.ylabel('Vibration')
plt.tight_layout()
```
